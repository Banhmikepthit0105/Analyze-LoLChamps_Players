```python
threshold = 0.4

# Tính tỉ lệ thiếu
missing_ratio_player = player_df.isnull().mean()
missing_ratio_champion = champion_df.isnull().mean()

# Giữ lại cột có tỉ lệ thiếu <= 40%
player_df = player_df.loc[:, missing_ratio_player <= threshold]
champion_df = champion_df.loc[:, missing_ratio_champion <= threshold]
```

```python
if 'As' in champion_df.columns:
    print("Column 'As' exists, proceeding with row deletion...")
    print("Initial DataFrame:")
    print(champion_df)
    print("\nInitial number of rows:", len(champion_df))

    # Step 2: Identify rows to drop
    # Condition 1: Rows with exactly 4 null values
    null_counts = champion_df.isna().sum(axis=1)
    rows_with_four_nulls = null_counts[null_counts >= 5].index

    # Condition 2: Rows where 'As' is null
    rows_with_null_as = champion_df[champion_df['As'].isna()].index

    # Combine both conditions (union of indices)
    rows_to_drop = rows_with_four_nulls.union(rows_with_null_as)

    # Step 3: Drop the identified rows
    champion_df = champion_df.drop(rows_to_drop)

    # Step 4: Display results
    print("\nRows dropped (4 nulls or null in 'As'):", len(rows_to_drop))
    print("\nFinal DataFrame:")
    print(champion_df)
    print("\nFinal number of rows:", len(champion_df))
else:
    print("Column 'As' does not exist in the DataFrame")
```

```python
# Bước 1: Group theo Year, Tournament, Team để tính tổng số trận thắng của tất cả thành viên trong đội
team_wins = player_df.groupby(['Year', 'Tournamment', 'Team']).agg({
    'W': 'sum'  # Tổng số trận thắng của tất cả thành viên trong đội
}).reset_index()

# Bước 2: Chia tổng số trận thắng của đội cho 5 để ra số trận thắng thực tế của đội
team_wins['team_wins'] = team_wins['W'] / 5

```

```python
# Bước 3: Group theo Year, Tournament để tính tổng số trận của mùa
season_games = team_wins.groupby(['Year', 'Tournamment']).agg({
    'team_wins': 'sum'  # Tổng số trận thắng của tất cả các đội = total_games
}).reset_index()
season_games.rename(columns={'team_wins': 'total_games'}, inplace=True)

```

```python
# Bước 4: Gộp total_games vào player_df
champion_df = champion_df.merge(
    season_games[['Year', 'Tournamment', 'total_games']],
    on=['Year', 'Tournamment'],
    how='left'
)
```

# Ý nghĩa các cột dữ liệu trong champion_stats

## Thông tin chung

- **Year**: Năm diễn ra giải đấu.  
- **Tournament**: Tên giải đấu (ví dụ: Spring Season).  
- **Champion**: Tên tướng được thống kê.  
- **Date**: Ngày xảy ra giải đấu  
- **Season**: Mùa giải (ví dụ: Spring, Summer).  
- **Event_Type**: Loại sự kiện (Có các giá trị như: Promotion, Season, Playoffs).  

## Thống kê về tướng (Champion)

- **G**: Số trận mà tướng xuất hiện (bao gồm cả khi được chọn hoặc bị cấm).  
- **PB**: Tỷ lệ chọn/cấm (Pick/Ban %) = (Số trận tướng xuất hiện / Tổng số trận trong mùa) * 100.  
- **B**: Số trận tướng bị cấm (Ban).  
- **GP**: Số trận tướng được chơi (Games Played).  
- **By**: Số người chơi đã sử dụng tướng này.  
- **W**: Số trận thắng của tướng.  
- **L**: Số trận thua của tướng.  
- **WR**: Tỷ lệ thắng (Win Rate) = (W / GP) * 100.  

## Hiệu suất trong trận

- **K**: Số mạng hạ gục trung bình (Kills).  
- **D**: Số lần bị hạ gục trung bình (Deaths).  
- **A**: Số lần hỗ trợ trung bình (Assists).  
- **KDA**: Tỷ lệ KDA = (K + A) / D.  
- **CS**: Số lính tiêu diệt trung bình (Creep Score).  
- **CS/M**: Số lính tiêu diệt mỗi phút (Creep Score per Minute).  
- **G.1**: Số vàng kiếm được trung bình (Gold).  
- **G/M**: Số vàng kiếm được mỗi phút (Gold per Minute).  

## Đóng góp vào đội

- **KPAR**: Tỷ lệ tham gia hạ gục (Kill Participation) = [(K + A) / Tổng số hạ gục của đội] * 100.  
- **KS**: Tỷ lệ đóng góp hạ gục (Kill Share) = (K / Tổng số hạ gục của đội) * 100.  
- **GS**: Tỷ lệ đóng góp vàng (Gold Share) = (G / Tổng số vàng của đội) * 100.  

## Vai trò

- **Mid Laner**: Tướng được chơi ở vị trí đường giữa (1 = Có, 0 = Không).  
- **Top Laner**: Tướng được chơi ở vị trí đường trên (1 = Có, 0 = Không).  
- **Bot Laner**: Tướng được chơi ở vị trí xạ thủ (1 = Có, 0 = Không).  
- **Jungler**: Tướng được chơi ở vị trí đi rừng (1 = Có, 0 = Không).  
- **Support**: Tướng được chơi ở vị trí hỗ trợ (1 = Có, 0 = Không).  

## Thống kê tổng quan

- **total_games**: Tổng số trận trong mùa giải.  
- **%B**: Tỷ lệ cấm của tướng = (B / total_games) * 100.  
- **%P**: Tỷ lệ chọn của tướng = (GP / total_games) * 100.

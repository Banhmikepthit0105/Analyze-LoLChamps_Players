```python
player_df = player_df.replace(['-', 'N/A', 'null', '[]'], np.nan)
champion_df = champion_df.replace(['-', 'N/A', 'null', '[]'], np.nan)
```

```python
# Chuyển các cột phần trăm và số từ object -> float
percent_cols = ['WR', 'KPAR', 'KS', 'GS', 'DMG']
for col in percent_cols:
    player_df[col] = player_df[col].str.replace('%', '', regex=False).str.replace('k', '', regex=False).replace('-', None).astype(float)

# Các cột số bị sai dtype
num_cols = ['G.1', 'G/M', 'DMG', 'DMG/M']
for col in num_cols:
    player_df[col] = pd.to_numeric(player_df[col], errors='coerce')
```

```python
percent_cols = ['PB', 'WR', 'KPAR', 'KS', 'GS', 'DMG']
for col in percent_cols:
    champion_df[col] = champion_df[col].str.replace('%', '', regex=False).str.replace('k', '', regex=False).replace('-', None).astype(float)

# Các cột bị sai kiểu
num_cols = ['K', 'B', 'By', 'D', 'A', 'KDA', 'CS', 'CS/M', 'G.1', 'G/M', 'DMG', 'DMG/M']
for col in num_cols:
    champion_df[col] = pd.to_numeric(champion_df[col], errors='coerce')
```

```python
threshold = 0.4

# Tính tỉ lệ thiếu
missing_ratio_player = player_df.isnull().mean()
missing_ratio_champion = champion_df.isnull().mean()

# Giữ lại cột có tỉ lệ thiếu <= 40%
player_df = player_df.loc[:, missing_ratio_player <= threshold]
champion_df = champion_df.loc[:, missing_ratio_champion <= threshold]
```

```python
# Điền NaN bằng giá trị trung bình và làm tròn mỗi cột của player_df
for col in player_df.select_dtypes(include='number').columns:
    player_df[col].fillna(player_df[col].mean(), inplace=True)
    player_df[col] = player_df[col].round(2)

# Điền NaN bằng giá trị trung bình và làm tròn mỗi cột của champion_df
for col in champion_df.select_dtypes(include='number').columns:
    champion_df[col].fillna(champion_df[col].mean(), inplace=True)
    champion_df[col] = champion_df[col].round(2)
```

```python
# Fill NaN còn sót lại bằng mean toàn bộ cột
for col in player_df.select_dtypes(include='number').columns:
    player_df[col].fillna(player_df[col].mean(), inplace=True)

for col in champion_df.select_dtypes(include='number').columns:
    champion_df[col].fillna(champion_df[col].mean(), inplace=True)
```

```python
champion_df['B'] = champion_df['B'].astype(int)
champion_df['By'] = champion_df['By'].astype(int)
```

```python
# Chuyển đổi cột 'As' từ chuỗi thành danh sách thực sự
champion_df['As'] = champion_df['As'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x
)

# Expand mỗi tướng thành nhiều dòng theo role
champion_df = champion_df.explode('As')
```

```python
champion_df
```

```python
# Bước 1: Group theo Year, Tournament, Team để tính tổng số trận thắng của tất cả thành viên trong đội
team_wins = player_df.groupby(['Year', 'Tournamment', 'Team']).agg({
    'W': 'sum'  # Tổng số trận thắng của tất cả thành viên trong đội
}).reset_index()

# Bước 2: Chia tổng số trận thắng của đội cho 5 để ra số trận thắng thực tế của đội
team_wins['team_wins'] = team_wins['W'] / 5

```

```python
# Bước 3: Group theo Year, Tournament để tính tổng số trận của mùa
season_games = team_wins.groupby(['Year', 'Tournamment']).agg({
    'team_wins': 'sum'  # Tổng số trận thắng của tất cả các đội = total_games
}).reset_index()
season_games.rename(columns={'team_wins': 'total_games'}, inplace=True)

```

```python
# Bước 4: Gộp total_games vào player_df
champion_df = champion_df.merge(
    season_games[['Year', 'Tournamment', 'total_games']],
    on=['Year', 'Tournamment'],
    how='left'
)
```

```python
season_max = player_df.groupby(['Year', 'Tournamment']).agg({
    'G/M': 'max',
    'CS/M': 'max',
    'KDA': 'max',
    'KPAR': 'max'
}).reset_index()

# Đổi tên cột để rõ ràng
season_max = season_max.rename(columns={
    'G/M': 'max_G/M',
    'CS/M': 'max_CS/M',
    'KDA': 'max_KDA',
    'KPAR': 'max_KPAR'
})

season_max['max_G/M'] = season_max['max_G/M'].fillna(0)
season_max['max_CS/M'] = season_max['max_CS/M'].fillna(0)
season_max['max_KDA'] = season_max['max_KDA'].fillna(0)
season_max['max_KPAR'] = season_max['max_KPAR'].fillna(0)

# Bước 2: Gộp các max value vào player_df
player_df = player_df.merge(
    season_max,
    on=['Year', 'Tournamment'],
    how='left'
)
```

```python
import random



# 1. Ánh xạ cột Tournamment sang Season (mùa) theo tiếng Việt, với xử lý đặc biệt cho năm 2023
def map_tournament_to_season(row):
    year = row['Year']
    tourn = row['Tournamment']
    # Xử lý riêng cho năm 2023
    if year == 2023:
        if tourn == 'Winter_Season':
            return 'Hừng Đông'
        elif tourn == 'Winter_Playoffs':
            return 'Hoàng Hôn'
    # Ánh xạ mặc định cho các trường hợp khác:
    mapping = {
        'Spring_Season': 'mùa Xuân',
        'Spring_Playoffs': 'mùa Xuân',
        'Spring_Promotion': 'mùa Xuân',
        'Summer_Season': 'mùa Hè',
        'Summer_Playoffs': 'mùa Hè',
        'Summer_Promotion': 'mùa Hè',
        'Winter_Season': 'mùa Đông',
        'Winter_Playoffs': 'mùa Đông'
    }
    return mapping.get(tourn, None)

champion_df['Season'] = champion_df.apply(map_tournament_to_season, axis=1)

# 2. Định nghĩa mapping khoảng thời gian (chuỗi ngày) cho từng cặp (Year, Season)
time_ranges = {
    (2018, 'mùa Xuân'): ('18/01/2018', '07/04/2018'),
    (2018, 'mùa Hè'): ('21/06/2018', '22/09/2018'),
    (2019, 'mùa Xuân'): ('17/01/2019', '13/04/2019'),
    (2019, 'mùa Hè'): ('20/06/2019', '15/09/2019'),
    (2020, 'mùa Xuân'): ('31/01/2020', '18/04/2020'),
    (2020, 'mùa Hè'): ('19/06/2020', '05/09/2020'),
    (2021, 'mùa Xuân'): ('22/01/2021', '18/04/2021'),
    (2021, 'mùa Hè'): (None, None),  # Mùa hè bị huỷ năm 2021
    (2021, 'mùa Đông'): ('16/11/2021', '26/12/2021'),
    (2022, 'mùa Xuân'): ('11/02/2022', '24/04/2022'),
    (2022, 'mùa Hè'): ('07/07/2022', '04/09/2022'),
    (2023, 'Hừng Đông'): ('24/02/2023', '23/04/2023'),
    (2023, 'Hoàng Hôn'): ('23/06/2023', '09/09/2023'),
    (2024, 'mùa Xuân'): ('20/01/2024', '07/04/2024'),
    (2024, 'mùa Hè'): ('21/06/2024', '18/08/2024')
}

def get_start_end_dates(row):
    key = (row['Year'], row['Season'])
    return pd.Series(time_ranges.get(key, (None, None)))

# Tạo các cột tạm chứa chuỗi ngày bắt đầu và kết thúc
champion_df[['Start_str', 'End_str']] = champion_df.apply(get_start_end_dates, axis=1)

# 3. Chuyển đổi chuỗi ngày sang kiểu datetime
champion_df['Start_Time'] = pd.to_datetime(champion_df['Start_str'], format='%d/%m/%Y', errors='coerce')
champion_df['End_Time']   = pd.to_datetime(champion_df['End_str'], format='%d/%m/%Y', errors='coerce')

# 4. Hàm lấy ngẫu nhiên một ngày trong khoảng [Start_Time, End_Time]
def pick_random_date(row):
    start = row['Start_Time']
    end = row['End_Time']
    if pd.isnull(start) or pd.isnull(end):
        return pd.NaT
    delta_days = (end - start).days
    random_offset = random.randint(0, delta_days)
    return start + pd.Timedelta(days=random_offset)

champion_df['Random_Date'] = champion_df.apply(pick_random_date, axis=1)

# 5. Thêm cột mới 'Data' vào champion_df chứa giá trị của Random_Date
champion_df['Date'] = champion_df['Random_Date']

# Nếu bạn không muốn giữ lại các cột tạm thời (Season, Start_str, End_str, Start_Time, End_Time, Random_Date),
# bạn có thể xóa chúng đi như sau:
champion_df.drop(columns=['Season', 'Start_str', 'End_str', 'Start_Time', 'End_Time', 'Random_Date'], inplace=True)

# Kiểm tra kết quả cuối cùng - champion_df có cột Data được thêm vào
print(champion_df)
```

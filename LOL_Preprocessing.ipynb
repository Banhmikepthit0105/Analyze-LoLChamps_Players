```python
import pandas as pd
import numpy as np
import ast
```

```python
player_df = pd.read_csv("player_stats.csv")
champion_df = pd.read_csv("champion_stats.csv")
```

```python
champion_df
```

```python
print('Players shape:', player_df.shape)
print('Champions shape:', champion_df.shape)
```

```python
player_df = player_df.replace(['-', 'N/A', 'null', '[]'], np.nan)
champion_df = champion_df.replace(['-', 'N/A', 'null', '[]'], np.nan)
```

```python
player_df.dtypes
```

```python
# Chuyển các cột phần trăm và số từ object -> float
percent_cols = ['WR', 'KPAR', 'KS', 'GS', 'DMG']
for col in percent_cols:
    player_df[col] = player_df[col].str.replace('%', '', regex=False).str.replace('k', '', regex=False).replace('-', None).astype(float)

# Các cột số bị sai dtype
num_cols = ['G.1', 'G/M', 'DMG', 'DMG/M']
for col in num_cols:
    player_df[col] = pd.to_numeric(player_df[col], errors='coerce')
```

```python
player_df.dtypes
```

```python
player_df
```

```python
champion_df.dtypes
```

```python
percent_cols = ['PB', 'WR', 'KPAR', 'KS', 'GS', 'DMG']
for col in percent_cols:
    champion_df[col] = champion_df[col].str.replace('%', '', regex=False).str.replace('k', '', regex=False).replace('-', None).astype(float)

# Các cột bị sai kiểu
num_cols = ['K', 'B', 'By', 'D', 'A', 'KDA', 'CS', 'CS/M', 'G.1', 'G/M', 'DMG', 'DMG/M']
for col in num_cols:
    champion_df[col] = pd.to_numeric(champion_df[col], errors='coerce')
```

```python
champion_df.dtypes
```

```python
missing_ratio_player = player_df.isnull().mean()
missing_ratio_player
```

```python
missing_ratio_champion = champion_df.isnull().mean()
missing_ratio_champion
```

```python
threshold = 0.4

# Tính tỉ lệ thiếu
missing_ratio_player = player_df.isnull().mean()
missing_ratio_champion = champion_df.isnull().mean()

# Giữ lại cột có tỉ lệ thiếu <= 40%
player_df = player_df.loc[:, missing_ratio_player <= threshold]
champion_df = champion_df.loc[:, missing_ratio_champion <= threshold]
```

```python
# Với player_df
for col in player_df.select_dtypes(include='number').columns:
    player_df[col] = player_df.groupby(['Year', 'Tournamment'])[col].transform(lambda x: x.fillna(x.mean()))

# Với champion_df
for col in champion_df.select_dtypes(include='number').columns:
    champion_df[col] = champion_df.groupby(['Year', 'Tournamment'])[col].transform(lambda x: x.fillna(x.mean()))
```

```python
print("Missing player_df:\n", player_df.isnull().mean() * 100)
```

```python
print("Missing champion_df:\n", champion_df.isnull().mean() * 100)
```

```python
# Fill NaN còn sót lại bằng mean toàn bộ cột
for col in player_df.select_dtypes(include='number').columns:
    player_df[col].fillna(player_df[col].mean(), inplace=True)

for col in champion_df.select_dtypes(include='number').columns:
    champion_df[col].fillna(champion_df[col].mean(), inplace=True)
```

```python
# Chuyển đổi cột 'As' từ chuỗi thành danh sách thực sự
champion_df['As'] = champion_df['As'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x
)

# Expand mỗi tướng thành nhiều dòng theo role
champion_df = champion_df.explode('As')
```

```python
champion_df
```

```python
# Bước 1: Group theo Year, Tournament, Team để tính tổng số trận thắng của tất cả thành viên trong đội
team_wins = player_df.groupby(['Year', 'Tournamment', 'Team']).agg({
    'W': 'sum'  # Tổng số trận thắng của tất cả thành viên trong đội
}).reset_index()

# Bước 2: Chia tổng số trận thắng của đội cho 5 để ra số trận thắng thực tế của đội
team_wins['team_wins'] = team_wins['W'] / 5

```

```python
# Bước 3: Group theo Year, Tournament để tính tổng số trận của mùa
season_games = team_wins.groupby(['Year', 'Tournamment']).agg({
    'team_wins': 'sum'  # Tổng số trận thắng của tất cả các đội = total_games
}).reset_index()
season_games.rename(columns={'team_wins': 'total_games'}, inplace=True)

```

```python
# Bước 4: Gộp total_games vào player_df
champion_df = champion_df.merge(
    season_games[['Year', 'Tournamment', 'total_games']],
    on=['Year', 'Tournamment'],
    how='left'
)
```

```python
champion_df
```

```python
player_df = player_df.merge(
    season_games[['Year', 'Tournamment', 'total_games']],
    on=['Year', 'Tournamment'],
    how='left'
)
```

```python
player_df
```

```python
# Bước 1: Tính %B và %P
champion_df['%B'] = (champion_df['B'] / champion_df['total_games']) * 100
champion_df['%P'] = (champion_df['GP'] / champion_df['total_games']) * 100

# Bước 2: Làm tròn kết quả đến 2 chữ số thập phân (tùy chọn)
champion_df['%B'] = champion_df['%B'].round(2)
champion_df['%P'] = champion_df['%P'].round(2)
```

```python
champion_df
```

```python
season_max = player_df.groupby(['Year', 'Tournamment']).agg({
    'G/M': 'max',
    'CS/M': 'max',
    'KDA': 'max',
    'KPAR': 'max'
}).reset_index()

# Đổi tên cột để rõ ràng
season_max = season_max.rename(columns={
    'G/M': 'max_G/M',
    'CS/M': 'max_CS/M',
    'KDA': 'max_KDA',
    'KPAR': 'max_KPAR'
})

season_max['max_G/M'] = season_max['max_G/M'].fillna(0)
season_max['max_CS/M'] = season_max['max_CS/M'].fillna(0)
season_max['max_KDA'] = season_max['max_KDA'].fillna(0)
season_max['max_KPAR'] = season_max['max_KPAR'].fillna(0)

# Bước 2: Gộp các max value vào player_df
player_df = player_df.merge(
    season_max,
    on=['Year', 'Tournamment'],
    how='left'
)
```

```python
player_df
```

```python
player_df['Champs'] = player_df['Champs'].apply(lambda x: ', '.join(ast.literal_eval(x)))
```

```python
player_df
```

```python
champion_df = champion_df[~champion_df['As'].isna()].reset_index(drop=True)

player_df = player_df[~player_df['Role'].isna()].reset_index(drop=True)
```
